```python
import pandas as pd
```

```python
schools = pd.read_csv('schools.csv')
schools.head()
```

```python
# filter best math schools
columns = ['school_name', 'average_math']

# find 80% of maximum score
threshold = 800 * 0.8 

best_math_scores = schools[schools['average_math']>= threshold]
best_math_schools = best_math_scores[columns].sort_values('average_math', ascending = False)
best_math_schools.head()
```

```python
# top 10 best performing schools across all 3 SAT scores

schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
columns = ['school_name', 'total_SAT']

top_10_schools = schools[columns].sort_values('total_SAT', ascending = False).head(10)
top_10_schools
```

```python
# compute std, mean all schools in the same borough
std_borough = schools.groupby('borough')['total_SAT'].std().round(2)
mean_borough = schools.groupby('borough')['total_SAT'].mean().round(2)
counts_borough = schools['borough'].value_counts()

# concatenate series in a dataframe
borough_stats = pd.concat([counts_borough, mean_borough, std_borough], axis = 1)

# rename columns
borough_stats.columns = ['num_schools', 'average_SAT', 'std_SAT']

# find borough with highest std_SAT value
borough_largest_std = borough_stats['std_SAT'].idxmax()

# dataframe with stats of borough with largest std_SAT
largest_std_dev = borough_stats.loc[[borough_largest_std]]
largest_std_dev
```
